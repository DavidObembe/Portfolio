```python
#Dependencies
import json
import pandas as pd
from pprint import pprint

#read in json file
file = "../Resources/ciafactbook.json"
data = open(file, encoding="utf8") 
  
# returns JSON object as  

factbook = json.load(data) 

```

```python
#import data from happiness csvs
```

```python
# create a list of all available countries
#change from all countries in factbook to all countries from happiness csvs
countrykeys = factbook['countries'].keys()
countries=[]
for country in countrykeys:
    countries.append(country)
countries
```

```python
#Loop for geo data
#variables for storing data
latitude = []
longitude = []
NSHemi = []
EWHemi = []
coastline = []
borders = []
coastpercent = []
elevations = []
forestpercent = []

for country in countries:
    try:
        latitude.append(factbook['countries'][f'{country}']['data']['geography']['geographic_coordinates']['latitude']['degrees'])
    except KeyError:
        print(f"{country} latitude KeyError")
        latitude.append("NaN")
    try:
        longitude.append(factbook['countries'][f'{country}']['data']['geography']['geographic_coordinates']['longitude']['degrees'])
    except KeyError:
        print(f"{country} longitude KeyError")
        longitude.append("NaN")
    try:
        NSHemi.append(factbook['countries'][f'{country}']['data']['geography']['geographic_coordinates']['latitude']['hemisphere'])
    except KeyError:
        print(f"{country} NSHemi KeyError")
        NSHemi.append("NaN")
    try:
        EWHemi.append(factbook['countries'][f'{country}']['data']['geography']['geographic_coordinates']['longitude']['hemisphere'])
    except KeyError:
        print(f"{country} EWHemi KeyError")
        EWHemi.append("NaN")
    try:
        coastline.append(factbook['countries'][f'{country}']['data']['geography']['coastline']['value'])
    except KeyError:
        print(f"{country} coastline KeyError")
        coastline.append("NaN")
    try:
        borders.append(factbook['countries'][f'{country}']['data']['geography']['land_boundaries']['total']['value'])
    except KeyError:
        print(f"{country} Borders KeyError")
        borders.append("NaN")
    try:
        elevations.append(factbook['countries'][f'{country}']['data']['geography']['elevation']['mean_elevation']['value'])
    except KeyError:
        print(f'{country} elevation key error')
        elevations.append('NaN')
    try:
        forestpercent.append(factbook['countries'][f'{country}']['data']['geography']['land_use']['by_sector']['forest']['value'])
    except KeyError:
        print(f"{country} forest KeyError")
        forestpercent.append("NaN")
    try:
        coastpercent.append(
                factbook['countries'][f'{country}']['data']['geography']['coastline']['value'] /
                (factbook['countries'][f'{country}']['data']['geography']['land_boundaries']['total']['value'] +
                factbook['countries'][f'{country}']['data']['geography']['coastline']['value']))
    except KeyError:
        coastpercent.append('NaN')
```

```python
#put geodata into a dataframe
geo_df = pd.DataFrame(list(zip(countries, latitude, longitude, NSHemi, EWHemi, coastline, borders, coastpercent, forestpercent, elevations)), 
               columns =['Country', 'Lat', 'Lng', 'NSHemi', 'EWHemi', 'coastline', 'borders', 'coastpercent','forestpercent', 'elevations'])
geo_df
```

```python
#loop for demographics data
#variables for storing data
dependencyratio= []
median_age = []
netmigrationrate = []
sexratioM2F = []
lifeexpentancy = []
fertilityrate = []

for country in countries:
    try:
        dependencyratio.append(factbook['countries'][f'{country}']['data']['people']['dependency_ratios']['ratios']['total_dependency_ratio']['value'])
    except KeyError:
        print(f"{country} dependencyratio key error")
        dependencyratio.append('NaN')
    try:
        median_age.append(factbook['countries'][f'{country}']['data']['people']['median_age']['total']['value'])
    except KeyError:
        print(f"{country} median_age KeyError")
        median_age.append("NaN")
    try:
        netmigrationrate.append(factbook['countries'][f'{country}']['data']['people']['net_migration_rate']['migrants_per_1000_population'])
    except KeyError:
        print(f"{country} netmigrationrate KeyError")
        netmigrationrate.append("NaN")
    try:
        sexratioM2F.append(factbook['countries'][f'{country}']['data']['people']['sex_ratio']['total_population']['value'])
    except KeyError:
        print(f"{country} sexratioM2F KeyError")
        sexratioM2F.append("NaN")
    try:
        lifeexpentancy.append(factbook['countries'][f'{country}']['data']['people']['life_expectancy_at_birth']['total_population']['value'])
    except KeyError:
        print(f"{country} lifeexpentancy KeyError")
        lifeexpentancy.append("NaN")
    try:
        fertilityrate.append(factbook['countries'][f'{country}']['data']['people']['total_fertility_rate']['children_born_per_woman'])
    except KeyError:
        print(f"{country} fertilityrate KeyError")
        fertilityrate.append("NaN")
```

```python
#create demographics dataframe
dem_df = pd.DataFrame(list(zip(countries, netmigrationrate, sexratioM2F, lifeexpentancy, fertilityrate, dependencyratio, median_age)), 
               columns =['Country', 'netmigrationrate', 'sexratioM2F', 'lifeexpentancy', 'fertilityrate', 'dependencyratio', 'median_age'])
dem_df.head()
```

```python
#loop for health and access data
#variables for storing data
cleanwateraccess = []
sanitationaccess = []
obesityrate = []
underweightchildren =[]

for country in countries:
    try:
        cleanwateraccess.append(factbook['countries'][f'{country}']['data']['people']['drinking_water_source']['improved']['total']['value'])
    except KeyError:
        print(f"{country} cleanwateraccess Key Error")
        cleanwateraccess.append('NaN')
    try:
        sanitationaccess.append(factbook['countries'][f'{country}']['data']['people']['sanitation_facility_access']['improved']['total']['value'])
    except KeyError:
        print(f"{country} sanitationaccess Key Error")
        sanitationaccess.append('NaN')
    try:
        obesityrate.append(factbook['countries'][f'{country}']['data']['people']['adult_obesity']['percent_of_adults'])
    except KeyError:
        print(f"{country} Obesity Key Error")
        obesityrate.append('NaN')
    try:
        underweightchildren.append(factbook['countries'][f'{country}']['data']['people']['underweight_children']['percent_of_children_under_the_age_of_five'])
    except KeyError:
        print(f"{country} underweightchildren KeyError")
        underweightchildren.append('NaN')
```

```python
#create health and access dataframe
health_df =  pd.DataFrame(list(zip(countries, cleanwateraccess, sanitationaccess, obesityrate, underweightchildren)), 
               columns =['Country', 'cleanwateraccess', 'sanitationaccess', 'obesityrate', 'underweightchildren'])
health_df
```

```python
#loop for govern and econ data
#variables for storing data
governmenttype = []
pppPc = []
grosssavings = []
unemployment = []
belowpoverty = []

for country in countries:
    try:
        governmenttype.append(factbook['countries'][f'{country}']['data']['government']['government_type'])
    except KeyError:
        print(f"{country} governmenttype Key Error")
        governmenttype.append('NaN')
    try: 
        pppPc.append(factbook['countries'][f'{country}']['data']['economy']['gdp']['per_capita_purchasing_power_parity']['annual_values'][0]['value'])
    except KeyError:
        print(f"{country} pppPc Key Error")
        pppPc.append('NaN')
    try:
        grosssavings.append(factbook['countries'][f'{country}']['data']['economy']['gross_national_saving']['annual_values'][0]['value'])
    except KeyError:
        print(f"{country} gross savings Key Error")
        grosssavings.append('NaN')
    try:
        unemployment.append(factbook['countries'][f'{country}']['data']['economy']['unemployment_rate']['annual_values'][0]['value'])
    except KeyError:
        print(f"{country} unemployment KeyError")
        unemployment.append("NaN")
    try:
        belowpoverty.append(factbook['countries'][f'{country}']['data']['economy']['population_below_poverty_line']['value'])
    except KeyError:
        print(f"{country} below poverty KeyError")
        belowpoverty.append("NaN")
```

```python
#create gov and econ df
gov_df =  pd.DataFrame(list(zip(countries, governmenttype, pppPc, grosssavings, unemployment, belowpoverty)), 
               columns =['Country', 'governmenttype', 'pppPc', 'grosssavings', 'unemployment', 'belowpoverty'])
gov_df
```

```python
#loop for education and utility data
#variables for storing data
literacy = []
schoolyears =[]
internetpercent = []
electricpercent = []

for country in countries:
    try:
        literacy.append(factbook['countries'][f'{country}']['data']['people']['literacy']['total_population']['value'])
    except KeyError:
        print(f"{country} literacy KeyError")
        literacy.append('NaN')
    try:
        schoolyears.append(factbook['countries'][f'{country}']['data']['people']['school_life_expectancy']['total']['value'])
    except KeyError:
        print(f"{country} school life Key Error")
        schoolyears.append('NaN')
    try:
        internetpercent.append(factbook['countries'][f'{country}']['data']['communications']['internet']['users']['percent_of_population'])
    except KeyError:
        print(f"{country} internetpercent KeyError")
        internetpercent.append("NaN")
    try:
        electricpercent.append(factbook['countries'][f'{country}']['data']['energy']['electricity']['access']['total_electrification']['value'])
    except KeyError:
        print(f"{country} electricity KeyError")
        electricpercent.append("NaN")
```

```python
#create education and util df
ed_df =  pd.DataFrame(list(zip(countries, literacy, schoolyears, internetpercent, electricpercent)), 
               columns =['Country', 'literacy', 'schoolyears', 'internetpercent', 'electricpercent'])
ed_df
```

```python
#combine dfs
final_df = pd.DataFrame(list(zip(countries, literacy, schoolyears, internetpercent, electricpercent, governmenttype, pppPc, grosssavings, unemployment, belowpoverty, cleanwateraccess, sanitationaccess, obesityrate, underweightchildren, netmigrationrate, sexratioM2F, lifeexpentancy, fertilityrate, dependencyratio, median_age, latitude, longitude, NSHemi, EWHemi, coastline, borders, coastpercent, forestpercent, elevations)),
                        columns =['Country', 'literacy', 'schoolyears', 'internetpercent', 'electricpercent', 'governmenttype', 'pppPc', 'grosssavings', 'unemployment', 'belowpoverty', 'cleanwateraccess', 'sanitationaccess', 'obesityrate', 'underweightchildren', 'netmigrationrate', 'sexratioM2F', 'lifeexpentancy', 'fertilityrate', 'dependencyratio', 'median_age', 'latitude', 'longitude', 'NSHemi', 'EWHemi', 'coastline', 'borders', 'coastpercent', 'forestpercent', 'elevations'])
```
