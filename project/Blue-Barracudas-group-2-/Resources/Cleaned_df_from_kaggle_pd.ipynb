```python
#import csv files
happy15_df= pd.read_csv('../Resources/Happy15.csv')
happy16_df= pd.read_csv('../Resources/Happy16.csv')
happy17_df= pd.read_csv('../Resources/Happy17.csv')
happy18_df= pd.read_csv('../Resources/Happy18.csv')
```

```python
happy18_df2= happy18_df.rename(columns= {'Country or region':'Country','Score':'Happiness Score','GDP per capita':'GDP per Capita',
                           'Social support':'Family', 'Healthy life expectancy':'Life Expectancy','Freedom to make life choices':'Freedom', 'Perceptions of corruption':
                           'Perceptions of corruption','Generosity':'Generosity','Overall rank':'Overall Rank'})
happy18_df2
happy18_df2['Year']= 2018
happy18_df2.head()
#list(happy18_df2.columns) 
#happy18_df2= happy18_df2= happy18_df2.drop(['Whisker.high','Whisker.low','Dystopia.Residual'], axis = 1)
#happy18_df2
```

```python
allFrames=[happy15_df2,happy16_df2,happy17_df2,happy18_df2]
jointFrames= pd.concat(allFrames)
jointFrames.head()
```

```python
countedJFrames=jointFrames.groupby('Country').count()
countedJFrames.head()
```

```python
#find all the countries that are either less than 4 or wrongly labeled
masked_JF= (countedJFrames['Year']< 4)
lessthan4y_df =countedJFrames[masked_JF]

lessthan4y_df.head()
```

```python
jointFrames=jointFrames.replace('Hong Kong S.A.R., China', 'Hong Kong')
jointFrames=jointFrames.replace('Northern Cyprus', 'North Cyprus')
jointFrames= jointFrames[jointFrames.Country != 'Somaliland Region']
jointFrames= jointFrames[jointFrames.Country!= 'Somaliland region']
jointFrames=jointFrames.replace('Trinidad & Tobago', 'Trinidad and Tobago')
jointFrames=jointFrames.replace('Taiwan Province of China', 'Taiwan')
cleanJointFrame=jointFrames

#mask_hongkong = (jointFrames['Country']== 'Hong Kong')
#check =jointFrames[mask_hongkong]
#check

```

```python
#first scatter plot
mean_jointframes= cleanJointFrame.groupby('Country').mean()[['Happiness Score','GDP per Capita','Family','Freedom','Generosity','Perceptions of corruption']]
mean_jointframes= pd.DataFrame(mean_jointframes)
mean_jointframes= mean_jointframes.reset_index()
mean_jointframes
```

```python
y= mean_jointframes['Happiness Score']
x= mean_jointframes['GDP per Capita']
plt.scatter(x,y, alpha= 0.5, label= 'country')
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.show()
```

```python
y= mean_jointframes['Happiness Score']
x_family= mean_jointframes['Family']
plt.scatter(x_family,y, alpha= 0.5, label= 'country')
plt.xlabel("Family")
plt.ylabel("Happiness Score")
plt.show()
```

```python
y= mean_jointframes['Happiness Score']
x_Generosity= mean_jointframes['Generosity']
plt.scatter(x_Generosity,y, alpha= 0.5, label= 'country')
plt.xlabel("Generosity")
plt.ylabel("Happiness Score")
plt.show()
```

```python
y= mean_jointframes['Happiness Score']
x_Freedom= mean_jointframes['Freedom']
plt.scatter(x_Freedom,y, alpha= 0.5, label= 'country')
plt.xlabel("Freedom")
plt.ylabel("Happiness Score")
plt.show()
```

```python
y= mean_jointframes['Happiness Score']
x_corruption= mean_jointframes['Perceptions of corruption']
plt.scatter(x_corruption,y, alpha= 0.5, label= 'country')
plt.xlabel("Perceptions of corruption")
plt.ylabel("Happiness Score")
plt.show()
```
