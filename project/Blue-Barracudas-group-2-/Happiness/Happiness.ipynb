```python
#Dependencies
import pandas as pd

#import csvs
file15 = "../Resources/Happy15.csv"
file16 = "../Resources/Happy16.csv"
file17 = "../Resources/Happy17.csv"
file18 = "../Resources/Happy18.csv"
file19 = "../Resources/Happy19.csv"

#
df15 = pd.read_csv(file15)
df16 = pd.read_csv(file16)
df17 = pd.read_csv(file17)
df18 = pd.read_csv(file18)
df19 = pd.read_csv(file19)
```

```python
# Add a year column
df15['Year'] = 2015
df16['Year'] = 2016
df17['Year'] = 2017
df18['Year'] = 2018
df19['Year'] = 2019
```

```python
#rename columns
df17.rename(columns={'Happiness.Rank': 'Happiness Rank', 'Happiness.Score': 'Happiness Score',
                     'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
                     'Health..Life.Expectancy.': 'Health (Life Expectancy)',
                     'Trust..Government.Corruption.':'Trust (Government Corruption)',
                     'Dystopia.Residual': 'Dystopia Residual'})
df18.rename(columns={'Overall rank': 'Happiness Rank', 'Country or region': 'Country', 'Score': 'Happiness Score',
                     'GDP per capita': 'Economy (GDP per capita)', 'Social support': 'Family',
                     'Healthy life expectancy': 'Health (Life Expectancy)', 'Freedom to make life choices':'Freedom',
                     'Perceptions of corruption': 'Trust (Government Corruption)'})
df19.rename(columns={'Overall rank': 'Happiness Rank', 'Country or region': 'Country', 'Score': 'Happiness Score',
                     'GDP per capita': 'Economy (GDP per capita)', 'Social support': 'Family',
                     'Healthy life expectancy': 'Health (Life Expectancy)', 'Freedom to make life choices':'Freedom',
                     'Perceptions of corruption': 'Trust (Government Corruption)'})
```
